# Pendulo Forzado Amortiguado 

La ecuación del péndulo forzado amortiguado es aceleración + amortiguamiento + Gravedad = Fuerza.
Que como equación se ve:


$\frac{d^{2}\theta}{dt^{2}}$ + $\mu\frac{d\theta}{dt}$ + $\omega^{2}sen(\theta)$ = 0

Ejercicio 2.- **Ecuación no lineal no forzada sin amortiguamiento**

**a).-** Para un ángulo inicial de 70 grados en reposo y una frecuencia natural unitaria, realiza una grafía donde compares la solución numérica del ángulo y su velocidad con aquella de la ecuación lineal usando las mismas condiciones iniciales.  

```python
#a) Problema 2
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

t=np.linspace(0,10,1000)

def D(y,t):
    th ,  w = y
    dydt=[w,-np.sin(th)]
    return dydt
y0=[1.22,0.0]
sol = odeint(D,y0,t)

plt.figure(figsize=(10,6))
    
plt.plot(t, sol[:, 0], 'b', label='theta(t)')
plt.plot(t, sol[:, 1], 'g', label='velocidad angular(t)')
plt.legend(loc='best')
plt.xlabel('t(s)')
plt.grid(True)
plt.grid(color= '0.5', linestyle = '--', linewidth = 1)
plt.show()
```

**b).-** Construye el espacio fase para una frecuencia natural unitaria, usando trayectorias que partan del reposo comn un ángulo inicial que vaya de -90 grados a 90 grados, en intervalos de 10 grados.Suma dos trayectorias que den vueltas, una encontra de las manesillas del reloj y otra a favor.     

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

t=np.linspace(-200,200,1000)

def D(y,t):
    th ,  w = y
    dydt=[w,-np.sin(th)]
    return dydt

y0=[(90*np.pi)/180,0.0]
y02=[1.2,0.0]
sol = odeint(D,y0,t)
sol2 = odeint(D,y02,-t)

plt.figure(figsize=(19,19))

plt.subplot(2,2,1)
plt.plot(sol[:, 0],sol[:, 1])
plt.legend(loc='best')
plt.grid(True)
plt.grid(color= '0.5', linestyle = '--', linewidth = 1)
plt.xlabel(r"$x ( s )$", fontsize = 12)
plt.ylabel(r"$y (rad)$", fontsize = 12)
plt.title('Y1',fontsize = 19, verticalalignment = 'baseline', horizontalalignment = 'center')

plt.subplot(2,2,2)
plt.plot(sol2[:, 0],sol2[:, 1])
plt.legend(loc='best')
plt.grid(True)
plt.grid(color= '0.5', linestyle = '--', linewidth = 1)
plt.xlabel(r"$x ( s )$", fontsize = 12)
plt.ylabel(r"$y (rad)$", fontsize = 12)
plt.title('Y2',fontsize = 19, verticalalignment = 'baseline', horizontalalignment = 'center')

plt.subplot(2,2,3)
plt.plot(sol[:, 0]+sol2[:, 0],sol[:, 1]+sol2[:, 1])
plt.legend(loc='best')
plt.grid(True)
plt.grid(color= '0.5', linestyle = '--', linewidth = 1)
plt.xlabel(r"$x ( s )$", fontsize = 12)
plt.ylabel(r"$y (rad)$", fontsize = 12)
plt.title('y1 + y2',fontsize = 19, verticalalignment = 'baseline', horizontalalignment = 'center')

plt.show()
```
